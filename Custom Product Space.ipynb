```python
# For d3plus.py
import sys
sys.path.append("./modules")
import d3plus2 as d3plus
```

```python
import pandas as pd
```

```python
df = pd.read_table("./sourcedata/list_for_mali.csv", sep=";")
df.head()
```

```python
df.name = df.name.astype(str)
```

```python
network = open("classifications/industrial_pspace.json").read()

ps = d3plus.ProductSpace(
    id="name",
    name="description",
    size={'value':'share', 'scale':d3plus.RawJavascript('d3.scale.sqrt()')},
    color="color",
    graph_data=network,
    presence="presence"
)
ps.draw(df)
```

```python
open("./test.html", "w+").write(ps.dump_html(df))
```
